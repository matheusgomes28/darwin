```python
%matplotlib inline
from utilities import *
from matplotlib import pyplot as plt
import numpy as np

# Get the image for conversion here
img = read_image("test.jpg", "BGR2GRAY")
img_blur = read_image("test_blur.jpg", "BGR2GRAY")

# Get the fourier of both
fft1 = np.fft.fft2(img)
fft2 = np.fft.fft2(img_blur)

# Shift them
fft1 = np.fft.fftshift(fft1)
fft2 = np.fft.fftshift(fft2)

# Get magnitude spectrums
mag1  = 20*np.log(np.abs(fft1))
mag2  = 20*np.log(np.abs(fft2))

plt.imshow(mag2, cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.savefig("mag2.jpg")
```
