```python
%matplotlib inline
import numpy as np
import sys, os, cv2
from utilities import *
import matplotlib.pyplot as plt

# Loading an image here
img = read_image("obama.jpg", "BGR2GRAY")


# Function to get each pixel value
# for the fourier function
def F(u, v, gray_img):
    
    intensity = gray_img/255
    
    N = intensity.shape[0] # Rows
    M = intensity.shape[1] # Cols

    s = 0
    for x in range(M):
        for y in range(N):
            s += intensity[y, x]*np.exp(-2j*np.pi*(u*x/M+v*y/N))
            
    return np.absolute(s)


dim = (220, 200)
resized = cv2.resize(img, dim)

F = np.fft.fft(resized)
print(np.min(F))
plt.imshow(np.absolute(F)/np.max(np.absolute(F)), cmap="gray")
```
