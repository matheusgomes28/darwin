## Histogram Equalisation

In this notebook the intensity histogram for a grayscale image will be produced and transformed in order to equalise the contrast of the image; hopefully giving a more uniform dataset.

Histogram equalisation is a way to distribute the pixel intensities within an image in a more uniform manner. The image's cumulative frequency plot can be used as to transform the intensities in the image in order to achieve this; the details of which will be described further on in the notebook.


#### The Intensity Histogram
The intensity histogram is calculated using the OpenCV function calcHist which takes parameters:
* Images - the image that the histogram is being calculated of, passed in a list form.
* Channels - a list of the dimension channels used to compute the histogram, here it will be [0] to represent grayscale.
* Mask - There is no mask applied here.
* histSize - an array of the histogram size in each dimension, 256 to represent the range of intensity for an image.
* ranges - the lower and upper bins of the histogram, since it is uniform.

The outcome of this function will be a histogram that shows the presence of different pixel intensities (ranging from 0 to 256, representing a fully black or white pixel respectively) and their frequencies within an image. The results of this process can be seen below on a test image of the back of the eye, where the x-axis is the intensity and the y-axis is the frequency.

```python
# imports
%matplotlib inline
from utilities import *
import numpy as np
import cv2
import math
from matplotlib import pyplot as plt
```

```python
# load the image
img = read_image("eye_black.TIF")

# convert to grayscale
gray_img = convert_spaces(img, "RGB2GRAY")
alt_img = convert_spaces(img, "RGB2GRAY")
a_img = convert_spaces(img, "RGB2GRAY")

# generate the intensity histogram for the image
hist = cv2.calcHist([gray_img],[0],None,[256],[0,256])

# plot the grayscale image and the histogram
fig = plt.figure(figsize=(15,5))
ax1 = plt.subplot(1, 2, 1)
ax1.set_axis_off()
ax1.set_title("Grayscale Image")
ax1.imshow(gray_img, cmap="gray")

ax2 = plt.subplot(1, 2, 2)
ax2.set_title("Intensity Histogram for the Grayscale Image")
ax2.bar(range(0, 256), hist)
fig.tight_layout()
plt.show()
```

#### Cumulative Frequency
The cumulative frequency array of an image is calculated using the numpy function cumsum, which has parameters:

 * a - an input array, here is it the histogram values that have been divided by the number of pixels in the image being used.
 * axis - the axis here is not specified so the cumsum is calculated over the flattened array.
 
By dividing each element in this array by the number of pixels in the image the cumulative probabiliy graph  can be plotted that has the pixel intensity range as the x-axis values and 0 to 1 as the y-axis values.

```python
# calculate the cumulative probability of the grayscale image
n_pixels = gray_img.shape[0] * gray_img.shape[1]
cf = np.cumsum(hist/n_pixels)

# plot picture and the cumulative probability graph
fig = plt.figure(figsize=(15,5))
ax1 = plt.subplot(1, 2, 1)
ax1.set_axis_off()
ax1.set_title("Grayscale Image")
ax1.imshow(gray_img, cmap="gray")

ax2 = plt.subplot(1, 2, 2)
ax2.set_title("Cumulative Probability Graph")
ax2.plot(cf)
```

#### Equalisation

To have equal contrast in an image, the cumulative probability plot should look like a straight diagonal line through the origin. To achieve this the cumulative probability at each pixel value in the image is multiplied by the maximum intensity value of the new image (here it will be 256 as we are working with grayscale still) and this then becomes the new intensity value for the pixel:

$$ cp_i * 256 = pi_i $$

Where $ cp_i $ is the cumulative probability at pixel $ i $ and $ pi_i $ is the new, transformed intensity at pixel $ i $.

This calculation can be seen in the code below, with the cumulative probability graph being recalculated to show the equalised plot. On the plot, it can be seen that the line produced is not perfectly straight and has jagged sections; but a perfectly straight line would indicate that all the intensities have been distributed evenly.

```python
# for each pixel the intensity becomes the cumulative probability of 
# that intensity multiplied by the maximum intensity
for row in range(gray_img.shape[0]):
    for col in range(gray_img.shape[1]):
            alt_img[row, col] = cf[gray_img[row, col]] * 256
            
# Make sure we round everything and put is as int
alt_img = np.round(alt_img).astype("uint8")
        
# generate the intensity histogram for the altered image
eq_hist = cv2.calcHist([alt_img],[0],None,[256],[0,256])

# calculate the cumulative probability of the altered grayscale image
eq_cf = np.cumsum(eq_hist/(alt_img.shape[0] * alt_img.shape[1]))
    
# plot picture and the cumulative probability graph
fig = plt.figure(figsize=(15,5))
ax1 = plt.subplot(1, 3, 1)
ax1.set_axis_off()
ax1.set_title("Original Grayscale Image")
ax1.imshow(gray_img, cmap='gray')

ax2 = plt.subplot(1, 3, 2)
ax2.set_axis_off()
ax2.set_title("Equalised Grayscale Image")
ax2.imshow(alt_img, cmap='gray')

ax3 = plt.subplot(1, 3, 3)
ax3.set_title("Equalised Cumulative Probability Graph")
ax3.plot(eq_cf)

fig.tight_layout()
plt.show()
```

## Histogram Matching

It can be seen that the equalised cumulative probability graph is not a perfect straight line throuhg the origin, but one can see that it could be with further work, indicating that the process has worked. But looking at the before and after images, this transformation process does not seem to give the best results; some vessels in the eye have become more prominant and easier to view but the bright optic nerve appears to have lost the detail within and directly around it, which is important in the detection process. 

To normalise the data set of images, the average histogram can be found from the collection of images. This can then be used to alter an images cumulative probability graph, through histogram matching; manipulating an images histogram graph to conform to a desired distribution. Instead of transforming the cumulative probability for each pixel by the maximum intensity, each value will be transformed by the equation of the new graph shape desired. 

The method chosen was implemented in the paper by Ana-Maria Sevcenco and Wu-Sheng Lu, Perfect histogram matching PCA for face recognition, where a histogram matching algorithm was used for face images in order to remove the lighting conditions dependencies. The algorithm is briefly described in the list below:

1. Let $I_k$ be the list of indices $(y,x)$ of where the pixels with intensity $k \in \{0,\dots,255\}$ are located. Note that $|I_k|=h(k)$ where $h(k)$ is the histogram value of $k$.
2. Let $I=\{I_0, I_2, \dots, I_{255}\}$ be the list containg all the individual histogram intensity lits.
3. Write $I=\{(y_1, x_2),\dots,(y_n,x_n)\}$ explicitly -- i.e ravel the list $I$ so all the indices are in order of intensity -- where $n$ is the total number of pixels in the image.
4. For a model intensity histogram function $m(k)$, perform the following iteratively:
    * Assign the first $h(0)$ indices in $I$ the intensity 0, i.e `image[I[0:h(0)]]=0`.
    * Assign the next $h(1)$ indices in $I$ the intensity 1 and so on.

It is hard to visualise the effect of the above algorithm on the original image, but it essentially 'hard' shifts pixels intensities to match the desired model histogram. I (Mat) can gladly explain the motivation behind this algorithm and provide some diagrams if necessary.

The cell below defines a function to get the histogram of an image with more than one channel, for example the RGB histograms. It is evaluated using the test image and then the last cell implements the algorithm mentioned above to match the histogram. 

** Note: The actual implementation is quite compact (about 10 lines), since I spent quite a while making sure that the most efficient numpy functions were used. I understand that it can be quite hard to understand what is happening, but if you gus have any questions just let me know and I will do my best to help. **

```python
%matplotlib inline
import cv2
import numpy as np
import utilities as ut
from matplotlib import pyplot as plt


## Funciton definitions to get the histograms ##
def get_histogram(image):
    """
    This function will return the histogram
    of the given image using the OpenCV
    calcHist function.
    
    Args: 
        image - numpy array representing the image.
        
    Returns:
        Numpy matrix representing the histogram of image.
        Note each row is the histogram of each components.
    """
    # Get all the channels
    dims = image.shape # dimension of image (H, W, D)
    n_channels = 1 if len(dims) < 3 else dims[-1]
    n_pixels = dims[0]*dims[1]
    
    # Set the parameters accordignly
    channels = [0]
    sizes = [256]
    ranges = [0, 256]
    mask = None
    
    # Separate image into different components
    if n_channels > 1:
        components = [image[:,:,i] for i in range(n_channels)]
    else:
        components = [image]
        
    # Helper function to apply calcHist to a given image
    helper = lambda x : cv2.calcHist([x], channels, mask, sizes, ranges)
    
    # Now return the stacked results
    return np.hstack([helper(comp) for comp in components])/n_pixels

# the graph line for the average cumulative frequency probability
def interp_cf(x, mean_cf):
    return np.interp(x, np.arange(start=0, stop=256), mean_cf)

# the inverse graph line
def inv_interp_cf(y, mean_cf):
    return np.interp(y,  mean_cf, np.arange(start=0, stop=256))


# Open the image to compute the cum. histogram of each component
image = ut.read_image("eye_black.tif", "BGR2RGB")
image = cv2.resize(image, (100,100))
histogram = get_histogram(image) # Get RGB histogram
cf = np.apply_along_axis(np.cumsum, 0, histogram)

fig = plt.figure(figsize=(10,2))
ax = plt.subplot(1,3,1)
ax.set_title("Red component")
ax.bar(range(0,256), cf[:,0], color="r")

ax = plt.subplot(1,3,2)
ax.set_title("Green component")
ax.bar(range(0,256), cf[:,1], color="g")

ax = plt.subplot(1,3,3)
ax.set_title("Blue component")
ax.bar(range(0,256), cf[:,2], color="b")
fig.tight_layout(); plt.show();
```

```python
## Open the image in gray scale (RGB)
image = ut.read_image("eye_black.tif", "RGB2GRAY")
image = cv2.resize(image, (250, 200), interpolation=cv2.INTER_AREA)
original_hist = get_histogram(image)

# Get the number numper of rows and columns in the image
# for calculating how many pixels there are later on..
rows, cols = image.shape

# The next array contains a list of 2D matrices (256), 
# containing False and True values representing whether 
# or not the current pixel has the corresponding intensity
I_array = [image==i for i in range(256)]

# In the next few lines the actual final form of I is
# obtained using np.where. Note that each row in I
# corresponds to an index in the image array.
I = [np.where(locs) for locs in I_array]
I = np.hstack(I).T # Get the primitive locations
                 # Note I.shape = (rows*cols, 2)
    

def transformation(image, I, model):
    """
    This function will transform the histogram of
    the image given so that it matches the model
    histogram passed.
    
    Arguments:
        image - Numpy array representing the image data.
        I - List of numpy arrays of intensity locations
            in primitive form, e.g I = [(i,j)]
        model - Numpy array representing the desired histogram.
    Returns:
        Numpy array with the new transformed images.
    """
    
    # Copy the image so we don't actuallly alter the original
    copy = image.copy()
    
    # For loop doing the transformation
    counter = 0 # Keep track of current index in I
    for i in range(256):
        N = int(model[i])
      
        # Assign N first pixels to corresponding 
        # intensity        
        if N > 0: # Check whether assignments are actually made here
            copy[np.split(I[counter:counter+N], 2, axis=1)] = i
            
        # Update counter so we start from correct place in I 
        counter += N
        
    # Returned the modified image 
    return copy

# Obtain the mean histogram acquired from a dataset scan
new_model = np.loadtxt("histogram_model.txt")
# Normalise the histogram so it matches the number of pixels
new_model = (new_model*cols*rows)/np.sum(new_model)

# Transform the image using the function above 
new_image = transformation(image, I, new_model)
new_hist = get_histogram(new_image) # Get the new hist


# Showing the image
figure = plt.figure(figsize=(10,10))

ax1 = plt.subplot(2,2,1)
ax1.set_title("Original Image") 
ax1.imshow(image, cmap="gray")

ax2 = plt.subplot(2,2,3)
ax2.set_title("Changed Image")
ax2.imshow(new_image, cmap="gray")

ax3 = plt.subplot(2,2,2)
ax3.set_title("Original Histogram")
ax3.plot(original_hist)

ax4 = plt.subplot(2,2,4)
ax4.set_title("Changed Histogram")
ax4.plot(new_hist)

fig.tight_layout()
plt.show()
```
