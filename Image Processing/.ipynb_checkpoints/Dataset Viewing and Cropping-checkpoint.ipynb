### Getting Bounding Rectangles for the Images

Something useful would be to get the bounding rectangle for an image in either dataset. This would allow us to crop the image and keep only the important eye bit. This might improve the histogram manipulation stuff (theoretically).

Multiple different concepts will be used in this process. Firstly, the image will be filteres using a blur filter (ideally a Gaussian filter to preserve the edges). This will remove all the noise found in the black part of the image which does not contain the eye.

The the image will be passed through the OpenCV canny edge algorithm, which uses a sobel filter to remove noise (not enough hence the previous Gaussian filter applied), then use partial derivatives $G'_x$ and $G'_y$ to find edge gradient magnitude $M(x,y)$ where the changes are occuring, the formula used to find the magnitude of the gradient is

$$
 M(x,y) = \sqrt{G'_x(x,y)^2 + G'_y(x,y)^2}
$$

Due to the high resolution of our images, this will most likely find many different edges, as opposed to just one continuous edge around the eye.

In order to "link" the fragmented edges, a dilate process will be used. This will thicken some edges, filling the circular eye shape as well as improving the edge lines. The new assigned pixel after the dilation, $D(x,y)$ is briefly summarised below

$$
D(x,y) = \text{Max}(\text{src}((x-t, y-t), (x+t, y+t)))
$$

Where $t$ is a suitably defined spacing constant.

```python
import cv2
import analysis as an
import filtering as ft
import numpy as np

# Get the first image of mesidor to test
mesi_images = get_images(mesidor1)
colour_img = ut.read_image(mesi_images[4], "BGR2RGB")
gray_img = ut.convert_spaces(colour_img, "RGB2GRAY")

# Pre filtering to remove black noise
gaussian = ft.gaussian_kernel(7, [0,0], 3)
#gray_img = ft.image_conv(gray_img, gaussian)

# Get the canny edge detection + dilate stuff
canny_img = cv2.Canny(gray_img, 10, 60, 1) # Detect the edges
canny_img = cv2.dilate(canny_img, np.ones((50,50)))


# Plotting stuff
fig = plt.figure(figsize=(10, 5))
ax1 = plt.subplot(1,2,1)
ax1.set_axis_off()
ax1.set_title("Original Image")
ax1.imshow(colour_img)
ax2 = plt.subplot(1,2,2)
ax2.set_axis_off()
ax2.set_title("Canny Detected Image")
ax2.imshow(canny_img, cmap="gray")
fig.tight_layout(); plt.show();
```

### Defining the Bounding Box from Canny Detection

The image on the cell above demonstrates the output of the Canny detected, dilated image. From the output, it is obvious that the points of interest in the canny detected image are the minima and maxima points. Therefore, the coordinates of the bounding rectangle, $(x_1, y_1)$ and $(x_2, y_2)$, are defined as

$$
x_1 = \text{argmin}_x (\mathrm{I})\\
y_1 = \text{argmin}_y (\mathrm{I})\\
x_2 = \text{argmax}_x (\mathrm{I})\\
y_2 = \text{argmax}_y (\mathrm{I})\\
$$

Where $\mathrm{I}$ is simply the matrix representing the canny detected image.

```python
def get_rect(edges):
    """
    This function will return a tuple
    representing the edges (diagonal) of the
    bounding box in the canny detected image 
    gien.
    
    Args:
        edges - numpy array representing the edges.
        must be bicolour only (0 and 255).
        
    Returns:
        (x_1, y_1, x_2, y_2) coordinates of rect.
    """
    
    # Get the blank/edge colours (assume edge > blank)
    blank_c = np.min(edges)
    edge_c = np.max(edges)
    
    # Np where to find coords of coords of edges
    rows, cols = np.where(edges==edge_c)
    
    # Return the coordinates (top left then bottom right)
    return (np.min(cols), np.min(rows), np.max(cols), np.max(rows))


# Acquire the rect coordinates for the image above
rect = get_rect(canny_img)
print(rect)
# Draw it on the original pic
cv2.rectangle(colour_img, rect[:2], rect[2:], (255, 0, 0), 10)

# Plot them here
fig = plt.figure(figsize=(10, 5)) 
ax1 = plt.subplot(1,2,1)
ax1.set_axis_off()
ax1.set_title("Detected Rectangle")
ax1.imshow(colour_img)
ax2 = plt.subplot(1,2,2)
ax2.set_axis_off()
ax2.set_title("Cropped Image")
ax2.imshow(colour_img[rect[1]:rect[3],rect[0]:rect[2]])
plt.show();
```
